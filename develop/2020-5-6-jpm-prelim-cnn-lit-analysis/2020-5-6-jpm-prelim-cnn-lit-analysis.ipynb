# Preliminary Analysis of CNN approach to CRC histology

These notes are based on the paper:

*Predicting survival from colorectal cancer histology slides using deep learning: A retrospective multicenter study* 

by Jakob Kather

This is a second iteration applying ML/DL to multi-class classification of CRC tissue sample, and a more-correlated-to-risk-ratio-than-the-gold-standard "deep stroma score" 

They achieved a ~95% accuracy on a 9-class classification of the tissue types of small patches of CRC tissue slides. They utilized a transfer from a over-the-counter CNN trained on ImageNet (VGG19). 

See slide deck for better 

## Motivation/Context/Background

"Precision oncology depends on stratification of cancer patients into different groups with different tumor genotypes, phenotypes, and clinical outcome."

"few if any digital pathology biomarkers have made their way into the clinic so far"

Two barriers:

- lack of annotated publically available data.
- lack of validation to clinically relevant situations with real world data from different institutions.
